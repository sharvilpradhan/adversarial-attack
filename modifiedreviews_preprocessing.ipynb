```python
# convering modreviews into text files to be fed into the model
# Importing stuff
import numpy as np
import pandas as pd
import os
import math
import re
```

```python
zomato= pd.read_csv('zomato_modreviews.csv', delimiter = ',', encoding = "utf-8")
# reviewsZ = zomato["mod_review"]
imdb = pd.read_csv('imdb_modreviews.csv',delimiter = ',', encoding = "utf-8")
# reviewsI = zomato["mod_review"]
```

```python
# How many postive and negative reviews
posZ = []
negZ = []
posI = []
negI = []
# Some of rating 3 are pos and some are neg. Creating a new section to change later if needed.
# amb stands for ambigious. Because as a collection it is ambigious.
ambZ = []
ambI = []
for index, row in zomato.iterrows():
    rat = float(row["Rating"])
    revZ = row["mod_review"]
    revZ = ''
#     negative review
    if rat < 3:
        negZ.append(revZ)
#     ambigious review
    elif rat == 3:
        ambZ.append(revZ)
#     positive review
    else:
        posZ.append(revZ)

for index, row in imdb.iterrows():
#     rat has if its classified as neg or pos
    rat = row["orig_sentiment"]
    revI = row["mod_review"]
#     negative review
    if rat == "neg":
        negI.append(revI)
# #     ambigious review
#     positive review
    else:
        posI.append(revI)
```
